# Widget Utilities

```python
dbutils.widgets.combobox(name="top_n", defaultValue="5", choices=["5","10","15"], label="Top N")
dbutils.widgets.dropdown(name="network", defaultValue="5G", choices=["2G","3G","4G","5G"], label="Technology Type")
dbutils.widgets.text(name="customer_name", defaultValue="Anonymous", label="Customer Name")
```

```python
print(dbutils.widgets.get(name="top_n"))
print(dbutils.widgets.get(name="network"))
print(dbutils.widgets.get(name="customer_name"))
```
