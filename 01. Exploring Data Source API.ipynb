# Custom Schema

```python
from pyspark.sql.types import *
USER_SCHEMA = StructType([
    StructField("id", IntegerType()),
    StructField("age", IntegerType()),
    StructField("gender", StringType()),
    StructField("designation", StringType()),
    StructField("salary", IntegerType())
])
df = spark.read.csv(
        path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.dat", 
        sep = "|",
        header = True,
        schema = USER_SCHEMA
    )
df.display()
```

# Handling Bad Records

1. PERMISSIVE (Default - Automatically col goes to corrupt_record)
2. DROPMALFORMED (Bad Record will be simply rejected)
3. FAILFAST (Even when 1 data is bad, Execution fails)

```python
spark.read.json( 
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/access_logs.json",
    mode = "PERMISSIVE"
). display()
```

```python
spark.read.json( 
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/access_logs.json",
    mode = "DROPMALFORMED"
). display()
```

```python
spark.read.json( 
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/access_logs.json",
    mode = "FAILFAST"
). display()
```

# DataFrame Writer API

```python
df = spark.read.csv( 
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv", 
    header = True, 
    inferSchema = True
)

df.write.save(
    format = "parquet",
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet",
)

spark.read.parquet(
     "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet"
).display()
```

# Modes
1. errorIfExists
2. OVERWRITE
3. APPEND
4. IGNORE

```python
df = spark.read.csv( 
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv", 
    header = True, 
    inferSchema = True
)

df.write.save(
    format = "parquet",
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet"
)

spark.read.parquet(
     "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet"
).display()
```

```python
df = spark.read.csv( 
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv", 
    header = True, 
    inferSchema = True
)

df.write.save(
    format = "parquet",
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet",
    mode = "OVERWRITE"
)

spark.read.parquet(
     "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet"
).display()
```

```python
df = spark.read.csv( 
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv", 
    header = True, 
    inferSchema = True
)

df.write.save(
    format = "parquet",
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet",
    mode = "APPEND"
)

spark.read.parquet(
     "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet"
).display()
```
