# Drawbacks of cloud

# Load employee.csv

```python
df = spark.read.csv(
  path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
  header = True,
  inferSchema = True
)

df.display()
```

# Writing the data as parquet

```python
df.write.format("parquet").save(
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet_02",
    mode = "ignore"
)

dataframe_parquet = spark.read.parquet(
    "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet_02"
)

dataframe_parquet.display()
```

# Drawbacks of Cloud
1. No update possible
2. Data quality issue exists
3. Overwrites the content

```python
dataframe_parquet.createOrReplaceTempView("temp_view")
```

```python
%sql
UPDATE temp_view
SET gender = 'F'
WHERE gender = 'Female'
```

```python
spark.read.csv(
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_006_new_column_education.csv",
    header = True,
    inferSchema = True
).write.format("parquet").mode("append").save(
    path = "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet_02"
)

spark.read.parquet(
    "/Volumes/merit_catalog/quickstart_schema/sandbox/output/output_parquet_02"
).display()
```
