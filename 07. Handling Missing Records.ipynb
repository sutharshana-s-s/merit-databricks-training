# Pre-Requisites

```python
df = spark.read.csv(
    path="/Volumes/merit_catalog/quickstart_schema/sandbox/dataset/employee.csv",
    header=True,
    inferSchema=True,
    sep="|",
    quote="'"
)
df.display()
```

# Handling Missing Values

## Using na.drop

```python
# Default
# Even if 1 valur of a row is null the whole row is dropped
df.na.drop(how="any").display()
```

```python
# Gets dropped when only all the values of the row is null
df.na.drop(how="all").display()
```

```python
# Gets dropped when either one of the columns have null
df.na.drop(subset=["id", "name"]).display()
```

```python
# Gets dropped when either one of the columns have null
df.na.drop(how="all", subset = ["id", "name"]).display()
```

## Using na.fill

```python
# Strings are saved as NULL IN SOURCE
df.na.fill("NULL IN SOURCE").display()
```

```python
# Strings are saved as NULL IN SOURCE and integer as -1
df.na.fill("NULL IN SOURCE").na.fill(-1, subset=["id"]).display()
```

```python
# Strings are saved as NULL IN SOURCE, integer as -1, Experience as 0
df.na.fill("NULL IN SOURCE").na.fill(-1, subset=["id"]).na.fill(0, subset=["exp"]).display()
```

## Using na.fill by storing the changing value as a list

```python
MISSING_VALUES_REPLACE = {
    "name":"Anonymous",
    "exp":0,
    "id":-1
}
df.na.fill(MISSING_VALUES_REPLACE).display()
```

## Filling average value

```python
from pyspark.sql.functions import avg

avg_exp = df.select(avg("exp")).collect()[0][0]
avg_exp
```

```python
MISSING_VALUES_REPLACE = {"name": "Anonymous", "exp": int(avg_exp), "id": -1}
df.na.fill(MISSING_VALUES_REPLACE).display()
```

## Fill DOB or DOJ

```python
from pyspark.sql.functions import when, current_date, col

df.withColumn(
    "doj", when(col("doj").isNull(), current_date()).otherwise(col("doj"))
).display()
```
